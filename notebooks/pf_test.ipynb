```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from sdes.particle_filter import simulate, particle_filter 
import sdeint
```

We have implemented a bootstrap particle filter for the linear, Gaussian SSM, which is defined by:
$$X_0 \sim \mathcal{N}(0, \sigma_x^2)$$
$$X_t | X_{t-1}=x_{t-1} \sim \mathcal{N}(x_{t-1}, \sigma_x^2)$$
$$Y_t | X_{t}=x_t \sim \mathcal{N}(x_t, \sigma_y^2) \qquad t=1, \dots, T$$

We have assumed a parameter of $\theta = (\sigma_x^2, \sigma_y^2)$

```python
# Test the particle filter:
T=100; N=100

params = {'sigma_x': 1.,
          'sigma_y': 1.,
          'sigma_x_0': 1.,
          }

X_true, data = simulate(T, params)
X, W, A = particle_filter(N, T, data, params)

test = X[:, 0:10] * W[:, 0:10]
```
