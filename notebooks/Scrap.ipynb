# Ito processes

Below, we write some code to simulate from the Ito process $dX_t = t \sin(t) dt + \sqrt{\theta} dW_t$, with $X_0 = 1$:

```python
def generate_W(T, n_points, theta):
    delta_t = T/(n_points - 1)
    t_s = np.linspace(0, T, num=n_points)
    W_increments = np.random.normal(scale=delta_t * theta, size=n_points); 
    W = W_increments.cumsum()
    return W
```

```python
f = lambda t: t * np.sin(t)
T = 10
theta = 10.
n_points = 1001

t_s = np.linspace(0, T, num=n_points)
W = generate_W(T, n_points, theta)
```

```python
fig, ax = plt.subplots()
for _ in range(20):
    ax.plot(t_s, f(t_s) + generate_W(T, n_points, theta))
ax.set_xlabel("t")
ax.set_ylabel("X_t")
ax.set_title("Simulation of an Ito process")
plt.show()
```
